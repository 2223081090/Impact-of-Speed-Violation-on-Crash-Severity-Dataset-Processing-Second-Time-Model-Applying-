```python
import csv
import random

# Set a seed for reproducibility (optional)
random.seed(42)

# Input and output file names
input_file = "Nigeria Crash Data P.csv"
output_file = "Nigeria Crash Data Shuffled.csv"

# Read all rows from the CSV
with open(input_file, mode='r', newline='', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    rows = list(reader)
    fieldnames = reader.fieldnames

# Extract each column's values into a separate list
columns = {col: [row[col] for row in rows] for col in fieldnames}

# Shuffle each column independently
for col in fieldnames:
    random.shuffle(columns[col])

# Rebuild rows from the shuffled columns
shuffled_rows = []
for i in range(len(rows)):
    new_row = {col: columns[col][i] for col in fieldnames}
    shuffled_rows.append(new_row)

# Write the shuffled data to a new CSV file
with open(output_file, mode='w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(shuffled_rows)

print(f"Shuffled dataset saved to {output_file}")
```
